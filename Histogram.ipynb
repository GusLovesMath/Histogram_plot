#### My general imports

````python
import matplotlib.pyplot as pl
import numpy as np

%config InlineBackend.figure_format = 'retina'
pl.rcParams.update( {'font.size': 15} )
````

Function we'll use to generate our random data
```python
sample_data = np.random.normal(mean, std, N)
```

````python
N = 100000
sample = np.random.normal( 10, 5, N )

mean = np.mean( sample )
std = np.std( sample )

print(f'random sample mean = {mean}')
print(f'random sample standard deviation = {std}')
````

The function we'll use to generate our histogram
```python
pl.hist( data, bins, color=None, label=None )
```

````python
bins = np.arange( min(sample), max(sample) + 1, 1)
pl.figure( figsize=( 8, 6 ) )

pl.hist( sample, bins, color='C9', edgecolor='k', linewidth=0.5)

pl.axvline(mean, c='r', label=f'mean = {mean:.2f}')
pl.axvline(mean - std, c='k', ls='--', label=f' std = {std:.2f}')
pl.axvline(mean + std, c='k', ls='--')
pl.title('Math & Physics Fun With Gus Histogram')
pl.ylabel('Occurences')
pl.xlabel('Value of Data')
pl.legend()
pl.grid( alpha=0.2 )
pl.savefig('hist.pdf')
pl.show()
````
